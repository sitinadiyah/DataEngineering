```python
import pandas as pd
import numpy as np
import re

df = pd.read_csv('data job posts.csv')
df['jobpost']
```

```python
#df['female'] = df['raw'].str.extract('(\d)', expand=True)
#pattern = re.compile(r'\w+(?=\sfox)')
#(\d+)([a-zA-Z]+)(\d+)
df['ApplicationDate'] = df['jobpost'].str.extract(re.compile(r"(\d\d [a-zA-Z]+ \d\d\d\d)"), expand=True)
```

```python
df['ApplicationDate']
```

```python
df['Duration'] = df['jobpost'].str.extract(re.compile(r"(\d [a-zA-Z]+)"), expand=True)
```

```python
df['Duration']
```

```python
df['title'] = df['jobpost'].str.extract(re.compile(r"(?s)TITLE\s*:\s*(.*)(?:$)"))
```

```python
df['title']
```

```python
df['Location'] = df['jobpost'].str.extract(re.compile(r"(?s)LOCATION:(.*)(?=\r)"))
```
